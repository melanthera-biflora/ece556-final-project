```python
import numpy as np
import pandas as pd
```

```python
#Import dataset (processed)
landsat_7 = pd.read_csv('/kaggle/input/mekong-delta-data/landsat7_df.csv')
landsat_8 = pd.read_csv('/kaggle/input/mekong-delta-data/landsat8_df.csv')
merged_data = pd.read_csv('/kaggle/input/mekong-delta-data/merged_data.csv')
salinity = pd.read_csv('/kaggle/input/mekong-delta-data/salinity_df.csv')
station = pd.read_csv('/kaggle/input/mekong-delta-data/station_with_id.csv')
```

```python
cols_to_drop = ['ST_B6', 'SR_B6', 'ST_B10']
merged_data = merged_data.drop(columns=cols_to_drop)
```

```python
merged_salinity = pd.merge(salinity, station, on='Station')
merged_salinity
```

```python
landsat7_mean = landsat_7[landsat_7['reducer'] == 'mean'].copy()
```

```python
# Create two separate DataFrames
landsat8_mean = landsat_8[landsat_8['reducer'] == 'mean'].copy()
```

```python
# Create two separate DataFrames
merged_mean = merged_data[merged_data['reducer'] == 'mean'].copy()
```

```python
# Convert 'date' columns to datetime (if exists)
for df in [landsat7_mean, landsat8_mean, merged_mean, merged_salinity]:
    df['date'] = pd.to_datetime(df['date'])
```

```python
# Split function
def split_by_year(df, year=2018):
    train = df[df['date'].dt.year < year].copy()
    test = df[df['date'].dt.year == year].copy()
    return train, test
```

```python
# Apply to each dataset
landsat7_train, landsat7_test = split_by_year(landsat7_mean)
landsat8_train, landsat8_test = split_by_year(landsat8_mean)
merged_train, merged_test = split_by_year(merged_mean)
salinity_train, salinity_test = split_by_year(merged_salinity)
```

```python
# Save Landsat 7
landsat7_train.to_csv('landsat7_train.csv', index=False)
landsat7_test.to_csv('landsat7_test.csv', index=False)

# Save Landsat 8
landsat8_train.to_csv('landsat8_train.csv', index=False)
landsat8_test.to_csv('landsat8_test.csv', index=False)

# Save merged reflectance + salinity
merged_train.to_csv('merged_train.csv', index=False)
merged_test.to_csv('merged_test.csv', index=False)

# Save salinity
salinity_train.to_csv('salinity_train.csv', index=False)
salinity_test.to_csv('salinity_test.csv', index=False)
```
