```python
import numpy as np
import pandas as pd
```

# Import Dataset

```python
#Import dataset
station = pd.read_csv('/kaggle/input/mekong-delta-data/station_with_id.csv')
salinity = pd.read_csv('/kaggle/input/mekong-delta-data/vietnam-mekong-data - salinity.csv')
landsat_5 = pd.read_csv('/kaggle/input/mekong-delta-data/landsat_5.csv')
landsat_7 = pd.read_csv('/kaggle/input/mekong-delta-data/landsat_7.csv')
landsat_8 = pd.read_csv('/kaggle/input/mekong-delta-data/landsat_8.csv')
```

## Salinity Preprocessing

```python
salinity.info()
```

```python
salinity_df = salinity.dropna()
```

```python
#Change object type to float64
def change_type(df, column, type1):
    if type1 == 'date':
        df[column] = pd.to_datetime(df[column]).dt.date
    else:
        df[column] = df[column].replace(r'[^\d.]', '', regex=True)  # remove non-numeric characters
        df[column] = pd.to_numeric(df[column], errors='coerce')
    return df[column]
```

```python
#Change object type for object column into float64
for column in ['5', '9','17', '19']:
    salinity_df[column] = change_type(salinity_df, column, 'float64')
salinity_df['Day'] = change_type(salinity_df, 'Day', 'date')
```

```python
salinity_df.info()
```

```python
salinity_df
```

```python
# Calculate row-wise statistics across hours
salinity_df['min'] = salinity_df.min(axis=1, numeric_only=True)
salinity_df['max'] = salinity_df.max(axis=1, numeric_only=True)
salinity_df['mean'] = salinity_df.mean(axis=1, numeric_only=True)
salinity_df['median'] = salinity_df.median(axis=1, numeric_only=True)
```

```python
salinity_df
```

```python
salinity_df = salinity_df.rename(columns={'Day': 'date'})
```

## Landsat Preprocessing

```python
#Note the source of the landsat dataframe
landsat_5['source'] = 'landsat_5'
landsat_7['source'] = 'landsat_7'
landsat_8['source'] = 'landsat_8'
```

```python
irrelevant_columns = [
    'QA_PIXEL', 'QA_RADSAT', 'SR_CLOUD_QA', 'SR_ATMOS_OPACITY',
    'ST_ATRAN', 'ST_CDIST', 'ST_DRAD', 'ST_EMIS', 'ST_EMSD',
    'ST_QA', 'ST_TRAD', 'ST_URAD', 'SR_QA_AEROSOL'
]
```

```python
landsat_5 = landsat_5.drop(columns=irrelevant_columns, errors='ignore')
```

```python
landsat_7 = landsat_7.drop(columns=irrelevant_columns, errors='ignore')
```

```python
landsat_8 = landsat_8.drop(columns=irrelevant_columns, errors='ignore')
```

```python
landsat_5['date'] = change_type(landsat_5, 'date', 'date')
landsat_7['date'] = change_type(landsat_7, 'date', 'date')
landsat_8['date'] = change_type(landsat_8, 'date', 'date')
```

```python
landsat_8.info()
```

```python
# Merge all three DataFrames
merged_df = pd.concat([landsat_5, landsat_7, landsat_8], ignore_index=True)
```
